```python
import rospy
from std_msgs.msg import Int64
from IPython.display import clear_output
from roboy_communication_middleware.msg import MotorCommand
from roboy_communication_middleware.msg import MotorStatus
#from roboy_communication_middleware.srv import MotorConfigService
```

```python
rospy.init_node('spinnaker_myoarm_interface')
rate = rospy.Rate(10) # 10 Hz input frequency
rospy.loginfo('starting-------------')
```

```python
def spinn_callback(data_input):
    # when readout_spikes are received --> send them as motor force 

    readout_spikes = data_input.data
    rospy.loginfo('spikes received-------------'+ str(readout_spikes))

    motor_forces = MotorCommand()
    motor_forces.motors = [12,13]
    #motor_forces.motors = [12]
    #motor_forces.setPoints [7,3]
    motor_forces.setPoints =[50000*readout_spikes/200, -50000*readout_spikes/200]
    
    
    pub_to_robot.publish(motor_forces)
    rospy.sleep(4)
    
```

```python
def robot_callback(data_input):
    # when new position values are received --> send them to the input neurons
    
    #msg-type: MotorStatus
    robot_state = data_input.position
    rospy.loginfo('robot state received-------------'+str(robot_state))

    
    input_spikes  = Int64()
    input_spikes.data = robot_state[12]/10000+5 # 13th motor
    
    pub_to_spinn.publish(input_spikes)
    rospy.sleep(4)
```

```python
rospy.Subscriber("/from_spinnaker", Int64, spinn_callback)
rospy.Subscriber("/roboy/middleware/MotorStatus", MotorStatus, robot_callback)
pub_to_spinn=rospy.Publisher("/to_spinnaker", Int64, queue_size=10)
pub_to_robot=rospy.Publisher("/roboy/middleware/MotorCommand", MotorCommand, queue_size=10)
```
